```python
import datetime
    
def doScrollDown(whileSeconds):
    start = datetime.datetime.now()
    end = start + datetime.timedelta(seconds=whileSeconds)
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(1)
        if datetime.datetime.now() > end:
            break
```

```python
doScrollDown(60)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
number = []
for item in soup.select('ul.calendar_list > li > a'):
    r = str(item).find('raffles')
    r2 = str(item).find('>\n')
    number.append(str(item)[r+8:r2-2])
```

```python
brand_list = []
for item in soup.select('div.product_info > div.brand'):
    brand_list.append(item.text.replace('\n', '').replace('  ', ''))
```

```python
product = []
for item in soup.select('div.product_info > div.name'):
    product.append(item.text.replace('\n', '').replace('  ', ''))
```

```python
# 발매처
place_count = []
for item in soup.select('div.product_info > div.detail_info'):
    r = item.text.replace('\n', '').find('조회수')
    place_count.append(item.text.replace('\n', '')[3:r])
```

```python
df = pd.DataFrame({'product_number':number, 'brand_k':brand_list, 'product_k':product, 'place_count':place_count})
```
